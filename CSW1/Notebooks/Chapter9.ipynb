```python
class GadiGhoda:
    def __init__(self, color, company, ttype):
        self.color = color
        self.company = company
        self.t = ttype
    def disp(self):
        print(f"Color:{self.color} \nCompany: {self.company} \nTransmission: {self.t}\n")

c1 = GadiGhoda("White", "Mahindra", "Manual")
print("Information of Car1")
c1.disp()
print("Information of Car2")
c2 = GadiGhoda("Black", "Tata", "Automatic")
c2.disp()
```

```python
class Number:
    def __init__(self, prefix):
        self.prefix= prefix
    def display(self, a,b,c):
        print(self.prefix, a, sep="")
        print(self.prefix, b, sep="")
        print(self.prefix, c, sep="")
printer = Number("-->")
printer.display(10,20,30)
```
