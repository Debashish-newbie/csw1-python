```python
while True:
    try:
        fname = input ("Enter the file to read")
        if not fname:
            break
        f = open(fname,'r')
        print(f.read())
        f.close()
        break
    except FileNotFoundError:
        print('File', fname, 'not found') #Executed when error will come
    else: #Execute when no error
        print(f.read())
        f.close()
        break
```
