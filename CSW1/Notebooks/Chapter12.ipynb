```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,label ='sin(x)')
plt.title("Sin Wave")
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
#Arange - create evenly space values within a range
#np.arrange(start, stop, step) ---- stop is excluded
arr = np.arange(0,10,0.5)
print(arr)
```

```python
#linspace ---- crt a array with evenly spaced no over a specified interval
#np.linspace(start, stop, num)
arr = np.linspace(0,1,5)
print(arr)
```

```python
#empty---- creates uninitialies array of specified space
#np.empty(shape, dtype = float)
arr = np.empty((2,3))
print(arr)
```

```python
#eye method ---- used to generate 2D idendity matrix
#np.eye(N, M= None, k=0, dtype = float)
arr = np.eye(3)
print(arr)
```

```python
#ones/zeros method --- creates an array filled with one
#np.ones(shape, dtype = float)
arr = np.ones((2,3))
print(arr)
arr = np.zeros((2,3))
print(arr)
```

```python
#full ---- initialise array with custom number
#np.full(shape, fill value, dtype = None)
arr = np.full((2,3),5)
print(arr)
```

```python
#fromfunction ---- creates an array by applying a func to each coordinate
#np.fromfunction(function, shape, dtype = float)
arr = np.fromfunction(lambda i, j: (i+1)*(j+1)  , (11,11))
print(arr)
```

```python
#Batch Operations on numpy arrays
#A+n: n will be added with each element of array 'A' same for A-n, A*n, A**n(expo)
import  numpy as np
arr = np.array([1,2,3,4])
result = arr*2
print("after multiplication with 2")
print(result)
r1 = 2**arr
print("Exponent")
print(r1)
print("Ground Division")
r2 =arr//2 #--- #floordiv
print(r2)
```

```python
#Another simple operation is to use arithematic operation on two numpy arrays
#The dimention of the two arrays must be same and the no of elements in each dimention should be same and the number of elements in each dimention should be same
import numpy as np

print("First Way")
arr1 =np.array( [[1,2,3,4],
        [4,2,1,5],
        [4,7,6,8],
        [5,7,3,9]])
print(arr1, "\n")

A = np.arange(16).reshape(4,4)
print("Second Way")
print(A,"\n")

B = np.fromfunction(lambda r, c: r*4 + c, (4,4), dtype=int)
print("Third way")
print(B,"\n")

I = np.identity(4, dtype=int)
print("Idendity Matix")
print(I,"\n")

print("Matrix")
C = A*1
print(C,"\n")

print("Squaring")
D = A*A
print(D,"\n")
print("Mult with idedity matrix")
E = A *I
print(E,"\n")
print("Mult with 10")
E = A*10
print(E,"\n")
```
