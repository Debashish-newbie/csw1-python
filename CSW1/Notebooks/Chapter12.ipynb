```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,10,100)
y = np.sin(x)
plt.plot(x,y,label ='sin(x)')
plt.title("Sin Wave")
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.show()
```

```python
import numpy as np
#np.array(object, dtype=None. int float)
lis = [1,2,3]
print(lis, type(lis))
arr = np.array(lis)
print(arr, type(arr))
```

```python
#Arange - create evenly space values within a range
#np.arrange(start, stop, step) ---- stop is excluded
arr = np.arange(0,10,0.5)
print(arr)
```

```python
#linspace ---- crt a array with evenly spaced no over a specified interval
#np.linspace(start, stop, num)
arr = np.linspace(0,1,5)
print(arr)
```

```python
#empty---- creates uninitialies array of specified space
#np.empty(shape, dtype = float)
arr = np.empty((2,3))
print(arr)
```

```python
#eye method ---- used to generate 2D idendity matrix
#np.eye(N, M= None, k=0, dtype = float)
arr = np.eye(3)
print(arr)
```

```python
#ones/zeros method --- creates an array filled with one
#np.ones(shape, dtype = float)
arr = np.ones((2,3))
print(arr)
arr = np.zeros((2,3))
print(arr)
```

```python
#full ---- initialise array with custom number
#np.full(shape, fill value, dtype = None)
arr = np.full((2,3),5)
print(arr)
```

```python
#copy--- creates deep copy of an array
#np.copy(a)
```

```python
#fromfunction ---- creates an array by applying a func to each coordinate
#np.fromfunction(function, shape, dtype = float)
arr = np.fromfunction(lambda i, j: (i+1)*(j+1)  , (11,11))
print(arr)
```

```python
rows = np.arange(1,11).reshape(-1,1)
cols = np.arange(1,11)
table = rows*cols
print(table)
```
