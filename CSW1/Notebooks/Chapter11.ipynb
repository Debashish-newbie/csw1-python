```python
import random as r
lis1, lis2, lis3 = [],[], []

for i in range (10):
    #normalvariate(mean , standard deviation)
    #High SD - Wide Spread, SD- low - Narrow Spread
    x = r.normalvariate(50, 10) #High prob in between 40-50-60
    lis1.append(int(x))
    
    #randint(start, end)  end is not included
    #Always generates integer
    y = r.randint(1,6)
    lis2.append(int(y))
    
    #random() generste a floating point number between 0 to 1 
    #but less than 1 not 1
    z = r.random()
    lis3.append((z))
    
 
    
print("Using normalvariate Distribution: ", lis1)
print("Using randint function: ", lis2)
print("Using random function: ", lis3)
    
```

```python
import random as r
#sample(poupulation , k) population - list, set ,tuple,etc
#chooses k random elements from the population 
#k should be less than len(population)
word = "PYTHON" #String
print(r.sample(word, 3))
    
fruits = ["apple", "banana", "cherry", "orange"] #List
print(r.sample(fruits, 2))

data = {1:"A", 2:"B", 3:"C", 4:"D"} #Dictionary
keys = r.sample(list(data.keys()),2)
print(keys)
vals = keys = r.sample(list(data.values()),2)

```
