```python
import random
coins = ["H", "T"]
coinslis ={}
h_count = 0
t_count = 0
for i in range (1000):
    t = random.choice(coins)
    if t == "H":
        h_count +=1
    elif t == "T":
        t_count +=1

print(f'Probability of Heads: {h_count/1000} Count: {h_count}')
print(f'Probability of Tails: {t_count/1000} Count: {t_count}')
    
```

```python
import random

coins = ["H", "T"]
tosses = [random.choice(coins) for _ in range(1000)]
freq = {coin: tosses.count(coin) for coin in coins}
for coin, count in freq.items():
    print(f'Probability of {coin}: {count/1000} Count: {count}')
```

```python
import random

coins = ["H", "T"]
freq = {coin: 0 for coin in coins}

for _ in range(1000):
    freq[random.choice(coins)] += 1

for coin, count in freq.items():
    print(f'Probability of {coin}: {count/1000} Count: {count}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of values from -15 to 20
a = np.linspace(-15, 20, 200)   # 200 points for smoothness
print(a)

# Define the function y = 3x - 5
y = 3*a - 5

# Plot the function
plt.plot(a, y, label="y = 3x - 5", color="blue")

# Add labels, title, legend, and grid
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Function Plot")
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
