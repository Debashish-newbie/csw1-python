```python
import random
coins = ["H", "T"]
coinslis ={}
h_count = 0
t_count = 0
for i in range (1000):
    t = random.choice(coins)
    if t == "H":
        h_count +=1
    elif t == "T":
        t_count +=1

print(f'Probability of Heads: {h_count/1000} Count: {h_count}')
print(f'Probability of Tails: {t_count/1000} Count: {t_count}')
    
```

```python
import random

coins = ["H", "T"]
tosses = [random.choice(coins) for _ in range(1000)]
freq = {coin: tosses.count(coin) for coin in coins}
for coin, count in freq.items():
    print(f'Probability of {coin}: {count/1000} Count: {count}')
```

```python
import random

coins = ["H", "T"]
freq = {coin: 0 for coin in coins}

for _ in range(1000):
    freq[random.choice(coins)] += 1

for coin, count in freq.items():
    print(f'Probability of {coin}: {count/1000} Count: {count}')
```

```python
import random
count = {x : 0 for x in range (2,13)}
print(count)
for i in range (1000):
    sum = 0
    for j in range (2):
        sum+= random.randint(1,6)
    count[sum]+=1
print("Frequencies:")
for i in count:
    print(f'{i} : {count[i]}')
        
```

```python
import random

freq_count = {i: 0 for i in range(10)}
for i in range(5000):
    r = random.randint(0, 9)
    freq_count[r] += 1
print(f"Num     Hits     Expected     Actual")
for k, v in freq_count.items():
    print(f"{k}       {v}         {1/10}       {round(v/5000, 3)}" )
```

```python
import random
stud_list = ["John", "Riya", "Meera", "Amit", "Rahul", "Zara","Sneha","Kunal"]
stud_sel =[]
for i in range (1,5):
    team = random.sample([i for i in stud_list if i not in stud_sel],2)
    print(f'Team{i}: {team}')
    stud_sel.extend(team)
    
```

```python
import random
freq_count = {x:0 for x in range (1,7)}
for i in range (2000):
    r = random.randint(1,6)
    freq_count[r]+=1
print(freq_count)
print(f"Num     Hits     Expected     Actual")
for k, v in freq_count.items():
    print(f"{k}       {v}         {1/10}       {round(v/2000, 3)}" )
print(sum([freq_count[x] for x in freq_count]))
print(f"Highest Occurances: {max(freq_count, key = freq_count.get)} ({freq_count[max(freq_count, key = freq_count.get)]})")
print(f"Lowest Occurances: {min(freq_count, key = freq_count.get)} ({freq_count[min(freq_count, key = freq_count.get)]})")
```
