```python
# write a function that given a word returns true if it mathes the pattern
# 1) Starts with c/C then  a vowel (a,e,i,o,u,,A,E,I,O,U)
# 2) Then any single character but not a vowel

import re

def matches_pattern(word):
    pattern = r'^[cC][aeiouAEIOU][^aeiouAEIOU]'
    return re.match(pattern, word) is not None

print(matches_pattern("cat"))
print(matches_pattern("dog"))
```

```python
#Write a regex with 3 groups to parse data in DD-MM-YYYYY . Match  the object to find whole match year span of whole match and last index

import re

pattern = re.compile(r"^(\d\d)-(\d\d)-(\d\d\d\d)$")
text = "23-05-2024"

m = pattern.match(text)

if m:
    print("Whole match:", m.group(0))
    print("Year:", m.group(3))
    print("Day:", m.group(1))
    print("Month:", m.group(2))
    print("Year span:", m.span(3))
    print("Last index of whole match:", m.end() - 1)
```
