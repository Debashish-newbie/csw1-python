```python
import sys
def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    if b == 0:
        print("Division by 0 not possible")
    else:
        print(a/b)

op = sys.argv[1]       
if op!= "exit":
    a,b =  int(sys.argv[2]), int(sys.argv[3])
    menu = {"add":add, "sub":sub, "mul":mul, "div":div, "exit":exit }
    menu[op](a,b)
else:
    print("Exiting")
```

```python
def fhs(x):
    for i in x:
        if i> 50:
            yield i
l= [25,80,45,120,60]
print(list(fhs(l)))
```

```python
import random
def sds():
    for i in range (10):
        yield random.randint(20, 30)
print(list(sds()))
```
