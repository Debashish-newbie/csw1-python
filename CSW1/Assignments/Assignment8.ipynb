```python
import numpy as np

v1 = np.random.randint(1, 10, size=20)
print("Vector v1:", v1)


max = np.max(v1)
min = np.min(v1)

print("Maximum value:", max)
print("Minimum value:", min)


count = 0
for x in v1:
    if x > min and x < max:
        count += 1
print("Count:", count )

count_between = np.sum((v1 > min) & (v1 < max)) #True = 1, False = 0
print("Number of elements between max and min:", count_between)





```
