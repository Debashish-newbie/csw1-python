```python
#Percentage sign format specifier
# %d and %i used for decimal
import math

print("%d (decimal form)" %42)
print("%i (decimal form)" %42)

# %f is for floating point number
print("%f (floating point form)" %105.868)

# %s is for String
print("%s (String form)" %"My name is Debashish")

# %r is for String----- Prints wih single quote
print("%r (String form)" %"My name is Debashish") 

# %x or %X is for Hexadecimal----- Prints decimal to hexadecimal
print("Hexadecimal of 50 is %x" %50)
# for i in range (0,16):
#     print("Hexadecimal of %d is %x" %(i,i))

# %o  is for Octal----- Prints decimal to octal
print("Octal of 50 is %o" %50)

# %g or %G is for Compact floating point number
g = 102.3456784
print("Compact form of %f is %g" %(g,g))

# %e or %E  is for representing in exponent form
print("Expoent form of 50 is %e" %50)
print("Expoent form of pi is %e" %math.pi)
print("Expoent form of 50 is %%e:%e" %50)

# %u is for representing nsigned integer
print("Unsignd integer form %d" %-65)
```

```python
print("Fruits are %s, %s, %s" %("apple", "banana", "orange"))
print("My name is %15s EOT" %"Debashish") #Add space chars including the string --- right justified
print("My name is %-15s EOT" %"Debashish") #Add space HYPEN chars including the string --- left justified
print("Right Justified %15d EOT" %100) #Add space chars including the string --- right justified
print("Left Justified %-15d EOT" %100) #Add space HYPEN chars including the string --- left justified
print("%.5d" %25) #Add preceeding 00 including num
print("%10.5d EOT" %25) #Add preceeding 00 including num
print("%-10.5d EOT" %25) #Add preceeding 00 including num
print("Fixed decimal %.3f" %2.34567643) #Prints till fixed decimal point
print("Fixed decimal %.3f" %2.34) #Prints till fixed decimal point adds zero to compensate
```

```python
#Instead of hardcoding numeric precision value space value in the format string we can pass the
# "*" as placeholdr and supply numeric value as argument
print("Number %5.3d" %250)
print("Number %*d" %(5 , 250)) #%(width, value)

print("Fixed decimal %10.3f" %2.34567643)
print("Fixed decimal %*.*f" %( 10, 3, 2.3456764))

print("My name is %15s EOT" %"Debashish")
print("My name is %*s EOT" %(15, "Debashish"))
print("My name is %*s EOT" %(-15 ,"Debashish"))
```
