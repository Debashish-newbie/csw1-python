```python
nums = [2,4,8,9,27,28]
prime, composite, square, cube = [], [], [], []
keys = ["Prime", "Composite", "Perfect Square", "Perfect Cube"]

for item in nums:
    p, c = True, False  # assume prime at start
    
    for i in range(2, int(item**0.5) + 1):
        if item % i == 0:
            c = True
            p = False
            break
    
    if p:
        prime.append(item)
    else:
        composite.append(item)
    
    root = int(item ** 0.5)
    if root * root == item:
        square.append(item)

    # --- Check perfect cube ---
    cube_root = round(item ** (1/3))
    if cube_root ** 3 == item:
        cube.append(item)

print(prime, composite, square, cube)
vals = [prime, composite, square, cube]
output = {keys[i]:vals[i] for i in range (4) }
op = {keys: l for keys, l in zip(keys, vals)}
print(output)
print(op)
```

```python
email = ['abc@gmail.com', 'xyz@hotmail.com', 'an.gmail.com', 'ery@gmail.com', 'fgt@yahoo.com', 'rty.soa@gmail.com', 'gty@hotmail.com']
fil = list(filter(lambda e : "@" in e and e.endswith(".com") or e.endswith(".org"), email))
dom = [x.split("@")[1].split(".")[0] for x in fil]
freq = {key : dom.count(key) for key in dom}
print("Valid Emails:", fil)
print("Domains:",set(dom))
print("Frequency:",freq)
```
