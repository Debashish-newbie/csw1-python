```python
#REGULAR EXPRESSION
#It deals with how to recognize pattern of words and characters
#So a RE is a pattern you specify using special characters to represent combinations of specified characters 
# or digits or words 
# 
#6.1 Introduction to RE -------------------------------------------------------------------------------
#RE is a word thats matches a given word eg RE- cat
#eg1 Match  a 'c' character
#eg2 Match any number of 'a' but atleast one
#ed Match a 't' character
#eg caaaa, caaaat, cat, caaaaaat , all above should be identified by RE
#eg ct - not identified
#eg ca+t +/0/* one or more

#In a RE two types of symbol are there 1) Literal 2) Special character---------------------------------------
#Literal - Chars must be matched exactly             


```

```python
import re 
pattern_phone = r'\d\d\d-\d\d\d-\d\d\d\d'
s = '943-465-8080'
if re.match(pattern_phone, s):
    print("number accepted")
else:
    print("number rejected")

s = '9473-465-8080'
if re.match(pattern_phone, s):
    print("number accepted")
else:
    print("number rejected")

s = '9a3-465-8080'
if re.match(pattern_phone, s):
    print("number accepted")
else:
    print("number rejected")
```

```python
#Refining Matches
import re 
pattern_phone = r'\d\d\d[-@]\d\d\d[-@]\d\d\d\d' #[@-] detect either @ or - anyone not both
s = '943-465-8080'
if re.match(pattern_phone, s):
    print("number accepted")
else:
    print("number rejected")

s = '943@465@8080'
if re.match(pattern_phone, s):
    print("number accepted")
else:
    print("number rejected")
```

```python
import re
part = r'(a+)(b+)(c+)'
m = re.match(part, "aabbccc")
print(m)
if m:
    print("hi")
print(m.group(0))  #Predefiined function prints total maching string
print(m.group(1))
print(m.group(2))
print(m.group(3))
for i in range (m.lastindex+1):
    print(m.group(i))

print(m.start(3))
print(m.end(2))
print(m.span(3))
```
