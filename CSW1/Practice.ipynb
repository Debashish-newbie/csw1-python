```python
exp = input("Enter a Expression:")

num=[]

for i in exp:
    for j in "+-/*":
        if(i==j):
            op=i
            num = exp.split(op)
            break

if(op == "+"):
    print("%d + %d = %d" %(int(num[0]), int(num[1]), int(num[0])+int(num[1])))
elif(op == "-"):
    print("%d - %d = %d" %(int(num[0]), int(num[1]), int(num[0])-int(num[1])))
elif(op == "*"):
    print("%d * %d = %d" %(int(num[0]), int(num[1]), int(num[0])*int(num[1])))
elif(op == "/"):
    print("%d / %d = %d" %(int(num[0]), int(num[1]), int(num[0])/int(num[1])))
```
